```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def alpha_e(n, alpha_c):
    """calculate probability of experimentwise error rate"""
    return (1 - (1 - alpha_c)**n)
```

```python
for i in range(1, 20):
    print('----',i,'----')
    print(alpha_e(n=i, alpha_c=0.05))
```

```python
x = range(1, 21)
y = [alpha_e(n=i, alpha_c=0.05) for i in x]
```

```python
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
plt.figure(figsize=(12,8))
sns.lineplot(x, y);
plt.gca().locator_params(nbins=11)
plt.xlabel('Number of Comparisons');
plt.ylabel('Pr(V>1)');
plt.title('Familywise Error vs. Number of Comparisons');
plt.savefig('fwer.pdf')
```
